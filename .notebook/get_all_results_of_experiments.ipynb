```python
import pandas as pd
import os
```

```python
path= os.getcwd().split('/')
i=path.index('Alignment-of-schema-only-and-instance-only-data')
folder_path= "/".join(path[:i+1])
gpt=pd.read_csv(folder_path + "/tests/outputs/All_results_gpt.csv")
gpt.head()
```

```python
llama=pd.read_csv(folder_path + "/tests/outputs/All_results_llama.csv")
llama.head()
```

```python
path_folder= folder_path + "/Results_Experiments/GPT_results"
gpt_execution=pd.DataFrame()
for name in ['ChEMBL','Magellan','TPC-DI','Wikidata']:
    df=pd.read_csv(os.path.join(path_folder,name+".csv"))
    gpt_execution=pd.concat([gpt_execution,df],axis=0)
    
gpt_execution.head()
```

```python
path_folder=folder_path + "/Results_Experiments/GPT_results"
llama_execution=pd.DataFrame()
for name in ['ChEMBL','Magellan','TPC-DI','Wikidata']:
    df=pd.read_csv(os.path.join(path_folder,name+".csv"))
    llama_execution=pd.concat([llama_execution,df],axis=0)

llama_execution.head()
```

```python
gpt.head()
```

```python
llama.columns
```

```python
for index, row in llama.iterrows():
    model, category, dataset_name = row["Model"], row["Category"], row["Dataset"]
    
    # Filtrer le DataFrame llama_execution
    filtered = llama_execution[
        (llama_execution["Category"] == category) &
        (llama_execution["Dataset"] == dataset_name) &
        (llama_execution["models"] == model)
    ]

    # Vérifier si "Execution_Time" existe
    if not filtered.empty:
        time = filtered.loc[:, "Execution_Time"].values[0]  # Récupérer la première valeur correspondante
        
        # Mettre à jour la valeur dans le DataFrame llama
        llama.loc[
            (llama["Category"] == category) &
            (llama["Dataset"] == dataset_name) &
            (llama["Model"] == model),
            "Execution_Time"
        ] = time
llama.head()
```

```python
for index, row in gpt.iterrows():
    model, category, dataset_name = row["Model"], row["Category"], row["Dataset Name"]
    
    # Filtrer le DataFrame gpt_execution
    filtered = gpt_execution[
        (gpt_execution["Category"] == category) &
        (gpt_execution["Dataset"] == dataset_name) &
        (gpt_execution["models"] == model)
    ]

    # Vérifier si "Execution_Time" existe
    if not filtered.empty:
        time = filtered.loc[:, "Execution_Time"].values[0]  # Récupérer la première valeur correspondante
        
        # Mettre à jour la valeur dans le DataFrame gpt
        gpt.loc[
            (gpt["Category"] == category) &
            (gpt["Dataset Name"] == dataset_name) &
            (gpt["Model"] == model),
            "Execution_Time"
        ] = time
gpt.head()
```

```python
llama.to_csv(folder_path + "/Results_Experiments/llama_results/llama_with_time.csv",index=False)
gpt.to_csv(folder_path + "/Results_Experiments/GPT_results/gpt_with_time.csv",index=False)
```
